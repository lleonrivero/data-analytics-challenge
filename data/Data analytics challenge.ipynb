# Understanding the data

```python
# Before cleaning the data, I will be adding a column for average consumption. 

data['avg_consumption'] = data['consume']/data['distance']
```

```python
df = data.pivot_table(index= ['gas_type'], values = ['avg_consumption', 'distance', 'speed', 'temp_inside', 'temp_outside'], aggfunc = {'avg_consumption':'mean', 'distance':'sum', 'speed':'mean', 'temp_inside':'mean', 'temp_outside':'mean'}).reset_index()
```

```python
df.head()
```

# Cleaning data

```python
# Standardizing header names

cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(' ', '_'))
data.columns = cols
data.columns
```

```python
# Removing duplicated rows, if any

data = data.drop_duplicates()
```

```python
data.head()
```

```python
# Replacing missing values in column temp_inside with the mean of the column

data['temp_inside'] = data['temp_inside'].fillna(np.mean(data['temp_inside']))
```

```python
data['temp_inside'].isna().sum()
```

```python
# Dropping columns:
    # 'specials': similar information provided in column 'ac'
    # 'refill_gas': information already provided in 'gas_type'

data = data.drop(['specials', 'refill_gas', 'refill_liters'], axis = 1)
```

```python
data.head()
```

```python
# Converting gas_type column with get_dummies function
```

```python
gas_type = pd.get_dummies(data['gas_type'])
data = pd.concat([data, gas_type], axis = 1)
data.head()
```

```python
# Dropping columns 'gas_type' and 'E10'

data = data.drop(['gas_type', 'E10'], axis = 1)
```

```python
# Rearranging column order

data.columns
```

```python
data = data[['distance', 'consume', 'speed', 'temp_inside', 'temp_outside', 'SP98', 'ac',
       'rain', 'sun', 'avg_consumption']]
data.head()
```

# Modelling: linear regression

## Xy Split

```python
X=data.drop(['avg_consumption'], axis=1)
y=data['avg_consumption']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
X_train.shape
```

### Scaling train set

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(X_train)
X_train_norm = transformer.transform(X_train)
X_train_norm = pd.DataFrame(X_train_norm, columns=X_train.columns)
X_train_norm.head()
```

### Scaling test set

```python
X_test_norm = transformer.transform(X_test)
X_test_norm = pd.DataFrame(X_test_norm, columns=X_test.columns)
X_test_norm.head()
```

## Running linear regression models

```python
from sklearn import linear_model

lm = linear_model.LinearRegression()
lm.fit(X_train_norm,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train_norm)
print('R2 score for the training set is:', r2_score(y_train, predictions))
```

```python
predictions_test = lm.predict(X_test_norm)
print('R2 score for the test set is:', r2_score(y_test, predictions_test))
```

```python
predictions_test
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```
